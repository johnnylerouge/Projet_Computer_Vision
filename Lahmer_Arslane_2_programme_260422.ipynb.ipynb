```python
# Fonction qui va retourner la race de chien la plus probable sur une image fournie

label_maps = {}
label_maps_rev = {}
for i, v in enumerate(train_dataset.class_names):
    label_maps.update({v: i})
    label_maps_rev.update({i : v})

```

```python
def predict_image(url, filename):
    os.system("curl -s {} -o {}".format(url, filename))
    img = Image.open(filename)
    img = img.convert('RGB')
    img = img.resize((180, 180))
    img.save(filename)
    plt.figure(figsize=(4, 4))
    plt.imshow(img)
    plt.axis('off')
    img = imread(filename)
    probs = model.predict(np.expand_dims(img, axis=0))
    for idx in probs.argsort()[0][::-1][:1]:
        print("la race de chien la plus probable est : ",  label_maps_rev[idx].split("-")[-1], '\n')
```
